```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
```

```python
df.dropna(inplace = True)
```

```python
x = df.drop(['prognosis'],axis = 1)
```

```python
x.corr()
```

```python
y = df['prognosis']
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RFclf = RandomForestClassifier(n_estimators = 50,max_depth = 10,random_state = 42)
```

```python
RFclf.fit(X_train,y_train)
```

```python
RFclf_score_ = accuracy_score(RFclf.predict(X_test),y_test)
```

```python
RFclf_score_
```

```python
RFclf_score = cross_val_score(RFclf,X_train,y_train,cv = 5)
```

```python
RFclf_score
```

```python
#testing on test data
test = pd.read_csv("Testing.csv")
xtest = test.drop(["prognosis"],axis = 1)
ytest = test["prognosis"]
```

```python
RFclf_score = accuracy_score(ytest,RFclf.predict(xtest))
```
